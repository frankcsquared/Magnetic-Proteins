```python
from magnets import Magnet_Space
import numpy as np
import random
```

```python
a = Magnet_Space(np.linspace(0, 40, 40), np.linspace(0,40,40)) 

for _ in range(10):
    a.add_magnet([random.randrange(-100,100), random.randrange(-100, 100)], [random.randrange(-20, 60), random.randrange(-20,60)])

a.recompute_field()

a.plot_magnitude()
a.plot_nulls()
a.plot_field_lines(normalized=False)
a.plot_stream()
```
